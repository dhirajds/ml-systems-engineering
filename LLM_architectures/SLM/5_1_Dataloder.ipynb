# Create a dataset with input data as input.

```python
from torch.utils.data import Dataset
import random
import torch
class SLM_Dataset(Dataset): # very important.inherits from Dataset.
    def __init__(self,  tokenizer, input_scentences = [], context_size = 4, stride = 3 ):
        self.input = input_scentences
        # convert the input in to tokens.
        self.tokenData =[]
        self.input_token_ids = []
        self.output_token_ids = []
        self.context_size =  context_size # how big is the max input.
        self.stride = stride # how much to slide for next input.
        for scentence in input_scentences:
            # scentence_tokens = tokenizer.encode(scentence)
            random_val = random.randint(5, 20)
            scentence_tokens = [ i for i in range(random_val, random_val+ 30)]

            self.tokenData.append(scentence_tokens)
            for i in range(0, len(scentence_tokens) - self.context_size - 1 , self.stride):
                # self.input_token_ids.append(scentence_tokens[i, i + self.context_size])
                # self.output_token_ids.append(scentence_tokens[i + 1 , i + self.context_size + 1])
                #while appending it should be torch .tensor
                self.input_token_ids.append(
                    torch.tensor(scentence_tokens[i: i + self.context_size]))
                self.output_token_ids.append(
                    torch.tensor(scentence_tokens[i + 1 : i + self.context_size + 1]))
    def __len__(self):
        return len(self.input_token_ids)
    def __getitem__(self, idx):
        print("get item input: ",self.input_token_ids[idx])
        print("get item output:",self.output_token_ids[idx])
        return self.input_token_ids[idx], self.output_token_ids[idx]
```

```python
from torch.utils.data import DataLoader
traindata = SLM_Dataset(tokenizer= "gpt", 
            input_scentences =["hi this is dhiraj"],
            context_size = 4,
            stride =3)
print(traindata.__len__())
trainDataLoader = DataLoader(traindata, batch_size= 3, shuffle =True)
```

```python
for x, y in iter(trainDataLoader):
    print("iter input : ",x.shape)
    print("iter output: ",y.shape)
```
